## 数位DP  
当我们需要枚举数字的时候 适合用数位DP  
相当于把一个数字看成一棵树 **相邻的更大数位是父节点**  
我们就可以应用记忆化搜索 从高位遍历到地位 再记忆化返回答案

灵神数位DP模板
```python
@cache
def f(i: int, mask: int, limit: int, valid: int):
    '''
    i: 当前填到第i位
    mask: 位运算记录当前已经填过的数字
    limit: 前面填的数字是否和n的前几位相同 这决定接下来能填的数字
    valid: 当前数字时候合法 (如果不限制数字使用次数的时候无所谓了)
    '''
    
    if i == len(s):  # 填满了 并且数字合法 返回记数(有时候不需要valid)
        return valid

    res = 0
    if not valid:    # 如果前面都是0 考虑继续填0
        res = f(i+1, mask, 0, 0)
    
    up = int(s[i]) if limit else 9   # 根据之前填的数字判断下一个数字的范围
    for d in range((1-valid), up+1): # 如果前面填了1 这里就填0 否则 从1开始填
        if mask >> d & 1 == 0:
            # 当且仅当 limit and d == up 才需要继续维持limit人设
            res += f(i+1, mask | 1<<d, limit and d == up, 1)
    return res

```

````python
# 经典题 统计特殊整数
class Solution:
    def countSpecialNumbers(self, n: int) -> int:

        # 每一位可以填0~9的任意数字 但是由于数字的特性 要做出一些限制
        s = str(n)

        # 返回从i开始构造数字 能构造的特殊整数的数目
        @cache
        def f(i: int, mask: int, limit: int, valid: int):
            '''
            i: 当前填到第i位
            mask: 位运算记录当前已经填过的数字
            limit: 前面填的数字是否和n的前几位相同 这决定接下来能填的数字
            valid: 当前数字时候合法 因为这题限制了数字使用的次数 为了前面可以一直填0(也就是不填数字) 
            '''
            # 填满了 返回isNum
            if i == len(s):
                return valid

            res = 0
            if not valid:  # 如果前面都是0 考虑继续填0
                res = f(i+1, mask, 0, 0)

            # 根据之前填的数字判断下一个数字的范围
            up = int(s[i]) if limit else 9
            # 如果前面填了1 这里就填0 否则 从1开始填
            for d in range((1-valid), up+1):
                if mask >> d & 1 == 0:
                    # 秒啊 当且仅当 limit and d == up 才需要继续维持limit人设
                    res += f(i+1, mask | 1<<d, limit and d == up, 1)
            return res

        return f(0, 0, 1, 0)
````

````python
class Solution:
    def atMostNGivenDigitSet(self, digits, n: int) -> int:
        
        s = str(n)

        # 这题必须用valid来往前面填0
        @cache
        def f(i, limit, valid):
            if i == len(s):
                return valid
            
            res = 0
            if not valid:
                res += f(i+1, 0, 0)
            
            up = s[i] if limit else '9'
            for d in digits:
                if d > up: break  # 有序
                res += f(i+1, limit and d==up, 1)

            return res
        
        return f(0, 1, 0)
````

````python
# 8月27日字节跳动第二次笔试 第四题 统计[a, b]之间数位最大值之和
a, b = map(int, input().split())
mod = 10**9 + 7
from functools import cache

# 可以不用valid 因为这题只关心最大数值
def solve(s):
    s = str(s)
    @cache
    def f(i, value, limit):
        if i == len(s):
            return value

        res = 0
        up = int(s[i]) if limit else 9
        for d in range(up+1):
            res += f(i+1, max(d, value), limit and d==up)

        return res
    
    return f(0, 0, 1);

print((solve(b)-solve(a-1)) % mod)
````

````python
mod = int(1e9 + 7)

def cal(b):
    a = 10
    res = (45 * b) % mod  # 45 = 1 + 2 + 3 + ... + 9
    b -= 1
    while b: # 快速幂
        if b & 1:
            res = (res * a) % mod
        a = (a * a) % mod
        b >>= 1
    return res

n = int(input())
print(cal(n // 2) * cal((n + 1) // 2) % mod)
````
